```python
#Face Detection in images and webcam


def image_detector():
    print("####################################################")
    print("####                OPTIONS MENU                ####")
    print("####################################################")
    print()
    print("  1.  Face Detection Already Saved Photo     ")
    print("  2.  Detect Face From Camera And Save photo ")
    print("  3.  Exit                                   ")
    print()
    
    try:
        ch=int(input("Choose Your Option :"))


        if ch==1:
            img=cv2.imread(input("Enter File name : "))
            img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)    

            gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            faceCascade = cv2.CascadeClassifier("C:\ProgramData\Anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")

            faces = faceCascade.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30),
            flags=0
            )
            print("no. of faces =",len(faces))
            # Draw a circle around the faces
            for (x, y, w, h) in faces:
                if w>h:
                    radius=w//2 #setting radius according to the size of the face
                else:
                    radius=h//2

                cv2.circle(img, (x+(w//2), y+(h//2)),radius, (0, 255, 0), 2) #(x+(w//2), y+(h//2))  => coordinates of center of face

            plt.imshow(img)

        elif ch==2:   
            faceCascade = cv2.CascadeClassifier("C:\ProgramData\Anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")

            video_capture = cv2.VideoCapture(0)
            i=0
            while True:
                # Capture frame-by-frame
                ret, frame = video_capture.read()

                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

                faces = faceCascade.detectMultiScale(
                    gray,
                    scaleFactor=1.1,
                    minNeighbors=5,
                    minSize=(30, 30),
                    flags=0
                )

                #print(frame.shape)
                cv2.putText(frame," Press 'q' or 'ESC' to EXIT    |     Press 's' to Click",(5,frame.shape[0]-25),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),1,cv2.LINE_AA,False)

                # Draw a circle around the faces
                for (x, y, w, h) in faces:
                    if w>h:
                        radius=w//2 #setting radius according to the size of the face
                    else:
                        radius=h//2

                    cv2.circle(frame, (x+(w//2), y+(h//2)),radius, (0, 255, 0), 2) #(x+(w//2), y+(h//2))  => coordinates of center of face

                    #print("val of x= ",x,"val of y=",y)

                cv2.imshow('WebCam', frame) 

                if (cv2.waitKey(1) & 0xFF == ord('q')) or (cv2.waitKey(1) & 0xFF == 27):
                    break

                elif cv2.waitKey(1) & 0xFF == ord('s'): #long press 's' to save 
                    cv2.imwrite('opencv_img_'+str(i)+'.jpg', frame)
                    i+=1
                    print(" Photo Clicked and Saved !!")
                    cv2.putText(frame," Photo Clicked ",(5,200),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),1,cv2.LINE_AA,False)
                    clicked=frame

                elif cv2.getWindowProperty('WebCam',1) == -1 : # To close window by close 'X' button or Alt+F4
                    break
            # When everything is done, release the capture
            video_capture.release()
            cv2.destroyAllWindows()
            clicked=cv2.cvtColor(clicked,cv2.COLOR_BGR2RGB)
            plt.imshow(clicked) # Display the last clicked image

        elif ch==3:
            pass
        else:
            print("Invalid Input...")

            
    except ValueError:
        print("Invalid Input....Aborting!!!")
    except Exception as e:
        print("Some error has occured, try again")

    
```
